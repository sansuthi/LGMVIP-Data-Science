### Problem Statement: 
**Read an image in RBG format and convert it to pencil sketch using cv2 library in Python.**

---

```python
# Importing Libraries
import cv2
import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"] = (10,5)
```

## Step 0: Loading the image

##### Reading image in RGB format and displaying it

```python
image = cv2.cvtColor(cv2.imread('image.png'), cv2.COLOR_BGR2RGB)
plt.title('ORIGINAL IMAGE')
plt.imshow(image)
plt.xticks([])
plt.yticks([])
plt.show()
```

##  Step 1: Converting image to Grayscale

##### Converting RBG image to a grayscale, to turn the image into a classic black and white image

```python
gray_image = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
plt.title('GRAYSCALE IMAGE')
plt.imshow(gray_image, cmap='gray')
plt.xticks([])
plt.yticks([])
plt.show()
```

## Step 2: Inverting grayscale Image
##### Invert the grayscale image, which gives the negative of the image

```python
inverted = 255 - gray_image
plt.imshow(inverted, cmap='gray')
plt.title('INVERTED GRAYSCALE IMAGE')
plt.xticks([])
plt.yticks([])
plt.show()
```

## Step 3: Blurring the inverted image
##### Converting the inverted grayscale image into a blurred one

```python
blurred = cv2.GaussianBlur(inverted, (21, 21), 0)
plt.imshow(blurred, cmap='gray')
plt.title('BLURRED IMAGE')
plt.xticks([])
plt.yticks([])
plt.show()
```

## Step 4: Inverting the blurred image

##### Now, inverting the blurred image of the inverted grayscale image

```python
invert_blurred = 255 - blurred
plt.imshow(invert_blurred, cmap='gray')
plt.title('INVERTED BLURRY IMAGE')
plt.xticks([])
plt.yticks([])
plt.show()
```

## Step 5: Inverted blurry image into pencil sketch  

##### Finally create the pencil sketch by mixing the grayscale image with the inverted blurry image, which can be done by dividing the images.

```python
pencil_sketch = cv2.divide(gray_image, invert_blurred, scale=250)
plt.imshow(pencil_sketch, cmap='gray')
plt.title('PENCIL SKETCH')
plt.xticks([])
plt.yticks([])
plt.show()
```

# Original image & Pencil sketch

```python
plt.figure(figsize=(18, 7))
plt.subplot(1, 2, 1)
plt.title('ORIGINAL IMAGE', fontsize=15)
plt.xticks([])
plt.yticks([])
plt.imshow(image, cmap='gray')
plt.subplot(1, 2, 2)
plt.title('PENCIL SKETCH', fontsize=15)
plt.xticks([])
plt.yticks([])
plt.imshow(pencil_sketch, cmap='gray')
plt.show()
```
